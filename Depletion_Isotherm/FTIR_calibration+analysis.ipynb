```python
#concentration of master solution for GMO standards between 8 mmol and 20 mmol.
GMO_RMM = 356.547 # g mol-1

Mass_GMO = ufloat(0.3075, 0.00005) #g
Volume_GMO = ufloat(0.010, (0.05/1000)) #dm-3
stockconc = (Mass_GMO/GMO_RMM)/Volume_GMO #mol dm-3

print(stockconc)
```

```python
# use a function to find the concentration of solutions after dilution stock solution.

def conc_after_dilution(stock_M, dod_M, stock_conc):
    dod_dens = ufloat(0.747, 0.003) #density of dodecane @ 25 deg C
    stock_dens = ufloat(0.745, 0.005) #measured in lab
    conc = (((stock_M/stock_dens)*stock_conc)/((stock_M/stock_dens)+(dod_M/dod_dens)))
    return conc
```

```python
# now we want to use regression to find the concentration of depletion isotherm samples.

def find_conc_from_standards(c, m, area):
    conc = (area - c)/m
    return conc

def add_element(dict, key, value):
    if key not in dict:
        dict[key] = []
    dict[key].append(value)
    
def tot_vol(stock, dod):
    dod_dens = ufloat(0.747, 0.003)
    stock_dens = ufloat(0.745, 0.005)
    return (((stock/stock_dens)+(dod/dod_dens))/1000)
```
