```python
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import norm
import pandas as pd
import numpy as np
import os
```

```python
#create some slabs that use the median values from the refl1d fit of the GMO data.
#it doesn't matter about up or down here so we'll stick with one.

SiO2_thickness = 1.831890E+01
Fe_thickness = 1.897273E+02
Fe_Up_SLD = 7.914021+(((7.914021*1E-6)/0.0000945)*2.091443E+00*0.00002699*1E6)
Fe_Down_SLD = 7.914021-(((7.914021*1E-6)/0.0000945)*2.091443E+00*0.00002699*1E6)
SiO2_Fe_roughness = 4.084097E+00
FeOx_thickness = 3.450993E+01
FeOx_Up_SLD = 5.637175E+00+1.148236E-01
FeOx_Down_SLD = 5.637175E+00-1.148236E-01
Fe_FeOx_roughness = 1.192139E+01
GMO_thickness = 1.938506E+01
GMO_solvation = 8.575932E+00*0.01
FeOx_GMO_roughness = 4.796275E+00
solvent_SLD = 6.319171E+00
GMO_solvent_roughness = 5.977790E+00


slabs = np.array([[0, 2.07, 0, 0, 0], #si
                  [SiO2_thickness, 3.47, 0, 3, 0], #sio2
                  [Fe_thickness, Fe_Up_SLD, 0, SiO2_Fe_roughness, 0], #Fe_up
                  [FeOx_thickness, FeOx_Up_SLD, 0, Fe_FeOx_roughness, 0], #feox_up
                  [GMO_thickness, 0.21, 0, FeOx_GMO_roughness, GMO_solvation], #GMO
                  [0, solvent_SLD, 0, GMO_solvent_roughness, 0]]) #solv
z = None
max_delta_z=None
```

```python
from refnx.reflect.interface import Interface, Erf, Step
f = Erf()
```

```python
#we need a distance for the original volume fraction and sld profile

nlayers = np.size(slabs, 0) - 2

max_delta_z = 0.5

# work on a copy of the input array
layers = np.copy(slabs)
layers[:, 0] = np.fabs(slabs[:, 0])
layers[:, 3] = np.fabs(slabs[:, 3])
# bounding layers should have zero thickness
layers[0, 0] = layers[-1, 0] = 0

# distance of each interface from the fronting interface
dist = np.cumsum(layers[:-1, 0])
zstart = -5 - 4 * np.fabs(slabs[1, 3])
zend = 5 + dist[-1] + 4 * layers[-1, 3]

# workout how much space the SLD profile should encompass
# (if z array not provided)
if z is None:
    npnts = 2500
    if max_delta_z is not None:
        max_delta_z = float(max_delta_z)
        npnts_deltaz = int(np.ceil((zend - zstart) / max_delta_z)) + 1

        npnts = max(200, npnts_deltaz)

    zed = np.linspace(zstart, zend, num=npnts)
else:
    zed = np.asfarray(z)
```

```python
sigma = layers[1:, 3]

Sivf = 1-f(zed, scale=sigma[0], loc=dist[0])
Sio2vf = f(zed, scale=sigma[0], loc=dist[0])-f(zed, scale=sigma[1], loc=dist[1])
Fevf = f(zed, scale=sigma[1], loc=dist[1])-f(zed, scale=sigma[2], loc=dist[2])
FeOxvf = f(zed, scale=sigma[2], loc=dist[2])-f(zed, scale=sigma[3], loc=dist[3])
GMOvf = f(zed, scale=sigma[3], loc=dist[3])-f(zed, scale=sigma[4], loc=dist[4])
solvvf = f(zed, scale=sigma[4], loc=dist[4])

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(zed, Sivf, label=r'Si_vf')
ax.plot(zed, Sio2vf, label=r'SiO2_vf')
ax.plot(zed, Fevf, label=r'Fe_vf')
ax.plot(zed, FeOxvf, label=r'FeOx_vf')
ax.plot(zed, GMOvf, label=r'GMOt_vf')
ax.plot(zed, solvvf, label=r'solv_vf')
ax.plot(zed, Sivf+Sio2vf+Fevf+FeOxvf+GMOvf+solvvf, label=r'Tot_vf')
ax.legend(frameon=False)
ax.set_ylabel(r'Volume Fraction')
ax.set_xlabel(r'Distance from Interface / $\mathrm{\AA{}}$')
```

```python
up_SLDs = np.array([2.07, 3.47, Fe_Up_SLD, FeOx_Up_SLD])
down_SLDs = np.array([2.07, 3.47, Fe_Down_SLD, FeOx_Down_SLD])

vol_fracs = np.array([Sivf, Sio2vf, Fevf, FeOxvf])

up_each_SLD = vol_fracs.T*up_SLDs
up_tot_SLD = np.sum(up_each_SLD, 1)

down_each_SLD = vol_fracs.T*down_SLDs
down_tot_SLD = np.sum(down_each_SLD, 1)

nuclear_SLD = (up_tot_SLD+down_tot_SLD)/2
magnetic_SLD = (up_tot_SLD-down_tot_SLD)/2
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(zed, up_tot_SLD)
ax[0].plot(zed, down_tot_SLD)

ax[1].plot(zed, magnetic_SLD)
ax[1].plot(zed, nuclear_SLD)
```

```python
#create arrays from csv data.

z = df_vf.values.T[0]
GMO = df_vf.values.T[1]*1E6
dodecane = df_vf.values.T[2]*1E6
```

```python
#modify the dodecane SLDs to match experimental median SLD.

dd_mod = ((solvent_SLD)/(np.average(dodecane[-16:-1])))*dodecane #scale dd sld
CMdod_mod = ((1.926268E+00)/(np.average(dodecane[-16:-1])))*dodecane #scale dd sld
hd_mod = ((-0.462)/(np.average(dodecane[-16:-1])))*dodecane #scale hd sld
```

```python
#extend the SLD profile of each component by taking last 15 data points in each array and repeating 10 times.
#this step is required so that the convolved data reaches a steady bulk SLD for solvent.

dd_mod = np.append(dd_mod, np.tile(dd_mod[-15:-1], 10))
CMdod_mod = np.append(CMdod_mod, np.tile(CMdod_mod[-15:-1], 10))
hd_mod = np.append(hd_mod, np.tile(hd_mod[-15:-1], 10))
GMO = np.append(GMO, np.tile(GMO[-15:-1], 10))

#extend the distance axis too.
z2 = np.append(z, z[-1]+np.cumsum(0.5*np.ones(dd_mod.shape[0]-z.shape[0])))

#then we can change the spacing of z2 for interpolation (improve convolution approximate to original SLD curve).
z2_interp = np.linspace(z2[0], np.round(z2[-1]*2)/2, int((np.round(z2[-1]*2)/2 - z2[0])/0.1)+1)
```

```python
GMO_ddod = GMO+dd_mod
GMO_CMdod = GMO+CMdod_mod
GMO_hdod = GMO+hd_mod

plt.plot(z2, GMO_ddod)
plt.plot(z2, GMO_CMdod)
plt.plot(z2, GMO_hdod)
```

```python
plt.plot(z2, GMO)
plt.plot(z2, hd_mod)
plt.plot(z2, GMO+hd_mod)
```

```python
#match spacing of zed to that of z2_interp.
zed2 = np.linspace(zed[0], np.round(zed[-1]*2)/2, int((np.round(zed[-1]*2)/2 - zed[0])/0.1)+1)
```

```python
#interpolate MD SLD profiles.

GMO_ddod_interp = np.interp(z2_interp, z2, GMO_ddod)
GMO_CMdod_interp = np.interp(z2_interp, z2, GMO_CMdod)
GMO_hdod_interp = np.interp(z2_interp, z2, GMO_hdod)
```

```python
#interpolate SLDn, SLDm and the vf profile of Si, SiO2, Fe, FeOx...

nuclear_SLD_interp = np.interp(zed2, zed, nuclear_SLD)
magnetic_SLD_interp = np.interp(zed2, zed, magnetic_SLD)

vf_interp = np.interp(zed2, zed, vol_fracs[0]+vol_fracs[1]+vol_fracs[2]+vol_fracs[3])
```

```python
#check SLDs look similar before and after interpolation

plt.plot(zed, magnetic_SLD)
plt.plot(zed, nuclear_SLD)

plt.plot(zed2, magnetic_SLD_interp)
plt.plot(zed2, nuclear_SLD_interp)
```

```python
#do the convolution here...

GMO_ddod_conv = np.convolve(conv_by, GMO_ddod_interp, mode='full')
GMO_CMdod_conv = np.convolve(conv_by, GMO_CMdod_interp, mode='full')
GMO_hdod_conv = np.convolve(conv_by, GMO_hdod_interp, mode='full')
```

```python
from refnx.dataset import ReflectDataset
from refnx.analysis import Transform, CurveFitter, Objective, GlobalObjective, Parameter, autocorrelation_chain, process_chain, integrated_time
from refnx.reflect import SLD, ReflectModel, MixedReflectModel, MixedSlab, Slab, Structure

#now define the models

#these are the scale and background parameters
#intensities
P1_GMO25dd_int = Parameter((9.483492E-01/2), 'P1_GMO25dd_int', vary=False, bounds=(0.5*0.8, 0.5*1.2))
P1_GMO25CMd_int = Parameter((9.398290E-01/2), 'P1_GMO25CMd_int', vary=False, bounds=(0.5*0.8, 0.5*1.2))
P1_GMO25hd_int = Parameter((9.470335E-01/2), 'P1_GMO25hd_int', vary=False, bounds=(0.5*0.8, 0.5*1.2))
#background
P1_GMO25dd_bkg = Parameter(4.850123E-06, 'P1_GMO25dd_bkg', vary=False, bounds=(1e-7, 2e-5))
P1_GMO25CMd_bkg = Parameter(9.209948E-06, 'P1_GMO25CMd_bkg', vary=False, bounds=(1e-7, 2e-5))
P1_GMO25hd_bkg = Parameter(6.061751E-06, 'P1_GMO25hd_bkg', vary=False, bounds=(1e-7, 2e-5))

#define adv_layer interface objective & global objective. These adv layers have their own SLD.
#models
P1_GMO25dd_M_conv = MixedReflectModel((P1_GMO25dd_U_alt, P1_GMO25dd_D_alt), scales=(P1_GMO25dd_int, P1_GMO25dd_int), bkg=P1_GMO25dd_bkg, dq=4.7096)
P1_GMO25CMd_M_conv = MixedReflectModel((P1_GMO25CMd_U_alt, P1_GMO25CMd_D_alt), scales=(P1_GMO25CMd_int, P1_GMO25CMd_int), bkg=P1_GMO25CMd_bkg, dq=4.7096)
P1_GMO25hd_M_conv = MixedReflectModel((P1_GMO25hd_U_alt, P1_GMO25hd_D_alt), scales=(P1_GMO25hd_int, P1_GMO25hd_int), bkg=P1_GMO25hd_bkg, dq=4.7096)
#objectives
P1_GMO25dd_obj_conv = Objective(P1_GMO25dd_M_conv, P1_GMO_ddod_25)
P1_GMO25CMd_obj_conv = Objective(P1_GMO25CMd_M_conv, P1_GMO_CMdod_25)
P1_GMO25hd_obj_conv = Objective(P1_GMO25hd_M_conv, P1_GMO_hdod_25)
#global obj
glob_objective_conv = GlobalObjective([P1_GMO25dd_obj_conv, P1_GMO25CMd_obj_conv, P1_GMO25hd_obj_conv])
```

```python
### median structure of fitted models ### 

#LAYERS#
Si_med = SLD(2.07, name='Si_med')
SiO2_med = SLD(3.47, name='SiO2_med')
Fe_med = SLD(7.914021E+00, name='Fe_med')
FeOx_med = SLD(5.637175E+00, name='FeOx_med')
GMO_med = SLD(0.21, name='GMO_med')

#SOLVENTS#
hdod_SLD_med = SLD(-0.462, name='hdod_med')
ddod_med = SLD(solvent_SLD, name='ddod_med')
CMdod_med = SLD(1.926268E+00, name='CMdod_med')

##SiO2##

SiO2_lay_med = SiO2_med(SiO2_thickness, 3)

SiO2_lay_med.thick.setp(vary=False, bounds=(1, 25))

##Fe##
# share the same Fe thickness, SiO2-Fe roughness - we specify variation of parameters here.
SiO2_Fe_R_med = Parameter(SiO2_Fe_roughness, 'SiO2_Fe_R_med', vary=False, bounds=(1, 15))
Fe_med.real.setp(vary=False, bounds=(7.5, 8.1))
Fe_thick_med = Parameter(Fe_thickness, 'Fe_t_med', vary=False, bounds=(150, 250))
Fe_magmom_med = Parameter(2.091443E+00, 'Fe_magmom_med', vary=False, bounds=(1.9, 2.2))
Fe_ScattLen_med = Parameter(0.0000945, 'Fe_SL_med', vary=False)

FeUp_lay_med = MixedMagSlabs2.USM_Slab(Fe_thick_med, Fe_med, Fe_magmom_med, Fe_ScattLen_med, SiO2_Fe_R_med)
FeDown_lay_med = MixedMagSlabs2.DSM_Slab(Fe_thick_med, Fe_med, Fe_magmom_med, Fe_ScattLen_med, SiO2_Fe_R_med)

##FeOx##
Fe_FeOx_R_med = Parameter(Fe_FeOx_roughness, 'FeOx_Fe_R_med', vary=False, bounds=(1, 15))
FeOx_med.real.setp(vary=True, bounds=(5.0, 7.2))
FeOx_thick_med = Parameter(FeOx_thickness, 'FeOx_t_med', vary=False, bounds=(20, 60))
FeOx_m_med = Parameter(1.148236E-01, 'FeOx_m_med', vary=False, bounds=(0, 1.3))

FeOxUp_lay_med = MixedMagSlabs2.USM_nomagmom_Slab(FeOx_thick_med, FeOx_med, FeOx_m_med, Fe_FeOx_R_med)
FeOxDown_lay_med = MixedMagSlabs2.DSM_nomagmom_Slab(FeOx_thick_med, FeOx_med, FeOx_m_med, Fe_FeOx_R_med)

### GMO

GMO_lay_med = GMO_med(GMO_thickness, 4.796275E+00)
GMO_lay_med.vfsolv.setp(GMO_solvation)

GMO_R = Parameter(GMO_solvent_roughness, 'GMO_R_med', vary=False, bounds=(1, 15))

### SOLVENT PARAMS FOR ADV_LAYER SOLVENT
#setup the solvent layers

hdod_lay_med = hdod_SLD_med(0, GMO_R)
ddod_lay_med = ddod_med(0, GMO_R)
CMdod_lay_med = CMdod_med(0, GMO_R)

GMO_dd_U = Si_med | SiO2_lay_med | FeUp_lay_med | FeOxUp_lay_med | GMO_lay_med | ddod_lay_med
GMO_dd_D = Si_med | SiO2_lay_med | FeDown_lay_med | FeOxDown_lay_med | GMO_lay_med | ddod_lay_med
GMO_CMd_U = Si_med | SiO2_lay_med | FeUp_lay_med | FeOxUp_lay_med | GMO_lay_med | CMdod_lay_med
GMO_CMd_D = Si_med | SiO2_lay_med | FeDown_lay_med | FeOxDown_lay_med | GMO_lay_med | CMdod_lay_med
GMO_hd_U = Si_med | SiO2_lay_med | FeUp_lay_med | FeOxUp_lay_med | GMO_lay_med | hdod_lay_med
GMO_hd_D = Si_med | SiO2_lay_med | FeDown_lay_med | FeOxDown_lay_med | GMO_lay_med | hdod_lay_med
```
