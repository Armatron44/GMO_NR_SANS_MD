```python
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import norm
import pandas as pd
import numpy as np
import os
```

```python
#list of median parameter values from the fit.

SiO2_thickness = 1.831890E+01
Fe_thickness = 1.897273E+02
Fe_Up_SLD = 7.914021+(((7.914021*1E-6)/0.0000945)*2.091443E+00*0.00002699*1E6)
Fe_Down_SLD = 7.914021-(((7.914021*1E-6)/0.0000945)*2.091443E+00*0.00002699*1E6)
SiO2_Fe_roughness = 4.084097E+00
FeOx_thickness = 3.450993E+01
FeOx_Up_SLD = 5.637175E+00+1.148236E-01
FeOx_Down_SLD = 5.637175E+00-1.148236E-01
Fe_FeOx_roughness = 1.192139E+01
GMO_thickness = 1.938506E+01
GMO_solvation = 8.575932E+00*0.01
FeOx_GMO_roughness = 4.796275E+00
solvent_SLD = 6.319171E+00
GMO_solvent_roughness = 5.977790E+00
```

```python
#load MD SLD profiles.

df = pd.read_csv(os.path.join(os.getcwd(), 'gmo_1_water_0_sld.csv'), sep=',' ,header=[0])
```

```python
# plot MD SLD data
%matplotlib inline
plt.plot(df.values.T[0], df.values.T[1], label='GMO')
plt.plot(df.values.T[0], df.values.T[2], label='dod')
plt.plot(df.values.T[0], df.values.T[3], label='tot')
plt.ylabel(r'SLD / $\mathrm{\AA}^{-2}\times{}10^{-6}$')
plt.xlabel(r'Distance from FeOx Surface / $\mathrm{\AA}$')
plt.legend()
```

```python
#modify SLD values of solvent to match those found in experiment fit (median values)

median_ddod_SLD = solvent_SLD
median_hdod_SLD = -0.462
median_CMdod_SLD = 1.926268E+00

#scale the dodecane profile
dd_mod = ((median_ddod_SLD)/(np.average(df.values.T[2][-15:])*1E6))*df.values.T[2]*1E6 #scale dd sld
hd_mod = ((median_hdod_SLD)/(np.average(df.values.T[2][-15:])*1E6))*df.values.T[2]*1E6 #scale hd sld
CMdod_mod = ((median_CMdod_SLD)/(np.average(df.values.T[2][-15:])*1E6))*df.values.T[2]*1E6 #scale hd sld

#plot to see how slds modified
plt.plot(df.values.T[0], dd_mod, label='dd')
plt.plot(df.values.T[0], hd_mod, label='hd')
plt.plot(df.values.T[0], CMdod_mod, label='CMdod')
plt.ylabel(r'SLD / $\mathrm{\AA}^{-2}\times{}10^{-6}$')
plt.xlabel(r'Distance from FeOx Surface / $\mathrm{\AA}$')
plt.legend()
```

```python
#load the SLD profiles from the median profile.
#these are inverted so that the midpoint of the solvent/GMO interface is 0.

medpp = np.loadtxt(os.path.join(os.getcwd(), 'P1-GMO-median-sim-1-0-profile.dat')).T
meddd = np.loadtxt(os.path.join(os.getcwd(), 'P1-GMO-median-sim-1-1-profile.dat')).T
plt.plot(medpp[0], medpp[1], label='Up')
plt.plot(meddd[0], meddd[1], label='Down')
plt.ylabel(r'SLD / $/mathrm{/AA}^{-2}/times{}10^{-6}$')
plt.xlabel(r'Distance from FeOx Surface / $/mathrm{/AA}$')
plt.legend()
```

```python
# find start of gradient from Fe to FeOx_R

cnstupslds = np.where(np.abs(np.diff(medpp[1])) < 1e-10)[0] #find SLDs that don't vary much.
upvalue = max(medpp[1][cnstupslds[0]:cnstupslds[-1]]) #find maximum value in this region of little difference.
startupvalindex = np.where(medpp[1] == upvalue)[0][0] #find position at where Fe to FeOx starts.

cnstdownslds = np.where(np.abs(np.diff(meddd[1])) < 1e-10)[0] #find SLDs that don't vary much.
downvalue = min(meddd[1][cnstupslds[0]:cnstupslds[-1]]) #find minimum value in this region of little difference.
startdownvalindex = np.where(meddd[1] == downvalue)[0][0] #find position at where Fe_R to FeOx_R starts.

# find the spin state that varies first.
if startupvalindex > startdownvalindex:
    pos_to_start = startupvalindex+1
else:
    pos_to_start = startdownvalindex+1
```

```python
# find distance from beginning of SLD profile to point at which Fe deviates from its median SLD value
distance_to_Fe_s = medpp[0][pos_to_start]-medpp[0][0]
print(distance_to_Fe_s)
```

```python
# Get nuclear and magnetic slds with resolution of 0.1 

Fe_sld_n = ((medpp[1][:pos_to_start]+meddd[1][:pos_to_start])/2)
Fe_sld_m = ((medpp[1][:pos_to_start]-meddd[1][:pos_to_start])/2)
```

```python
plt.plot(medpp[0][:pos_to_start], Fe_sld_n)
```

```python
#find location of FeOx-Fe interface
#if the midpoint of solvent/GMO interface is z = 0, then Fe_FeOx midpoint is
#GMO thickness + FeOx thickness.

Fe_FeOx_mdpoint = GMO_thickness + FeOx_thickness
```

```python
# get slds to midpoint of GMO/FeOx interface (or GMO thickness):

if np.round((medpp[0][np.where(medpp[0] > GMO_thickness)[0][0]] - GMO_thickness)*10)/10 == 0.1:
    FeOx_int_ind = np.where(medpp[0] > GMO_thickness)[0][0]-1
else:
    FeOx_int_ind = np.where(medpp[0] > GMO_thickness)[0][0]
```

```python
#now we modify the sldn and sldm of FeOx slds that fall below the median sldn and sldm values so that they equal
#the median sldn and sldm values.
Fe_FeOx_sldn = np.where(Fe_sld_n[FeOx_int_ind:] < 5.637175E+00, 5.637175E+00, Fe_sld_n[FeOx_int_ind:])
Fe_FeOx_sldm = np.where(Fe_sld_m[FeOx_int_ind:] < 1.148236E-01, 1.148236E-01, Fe_sld_m[FeOx_int_ind:])
```

```python
#now take these modified Fe_FeOx sld profiles and stitch the MD sld profiles on to them.
#MD sld profiles are in 0.5 , so if we take every 5th SLD value from the modified Fe_FeOx sld profiles, we should match
#delta Z.

comp_sldm = np.insert(Fe_FeOx_sldm[::5], 0, np.zeros(np.flip(dd_mod + df.values.T[1]*1E6).shape[0]))

dd_comp_sldn = np.insert(Fe_FeOx_sldn[::5], 0, np.flip(dd_mod + df.values.T[1]*1E6))
hd_comp_sldn = np.insert(Fe_FeOx_sldn[::5], 0, np.flip(hd_mod + df.values.T[1]*1E6))
CMdod_comp_sldn = np.insert(Fe_FeOx_sldn[::5], 0, np.flip(CMdod_mod + df.values.T[1]*1E6))
```

```python
plt.plot(dd_comp_sldn, label='dd')
plt.plot(hd_comp_sldn, label='hd')
plt.plot(CMdod_comp_sldn, label='CMdod')
plt.ylabel(r'SLD / $\mathrm{\AA}^{-2}\times{}10^{-6}$')
plt.xlabel(r'Distance / $\mathrm{\AA}$')
plt.legend()
```

```python
%matplotlib inline

fig, ax = plt.subplots()

#does stitching affect the sld profile?
ax.plot(np.flip(dd_comp_sldn), color='blue', label='stitched')

#align original MD SLD profile to the end of our stitched SLD.
align_factor = len(np.flip(dd_comp_sldn))-len(dd_mod + df.values.T[1]) 

ax.scatter(np.linspace(0, len(dd_mod + df.values.T[1])-1, len(dd_mod + df.values.T[1]))+align_factor, 
           (dd_mod + df.values.T[1]*1E6), color='red', linestyle='-.', label='Original')

ax.set_xlim(210, 300)
ax.set_ylim(-0.5, 7)
ax.set_ylabel(r'SLD / $\mathrm{\AA{}}^{-2} \times{} 10^{-6}$')
ax.set_xlabel(r'Distance / $\mathrm{\AA{}}$')

ax.legend(frameon=False)
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.219*2.83465)
ax.xaxis.set_tick_params(width=0.219*2.83465)
ax.yaxis.set_tick_params(width=0.219*2.83465)
ax.xaxis.set_tick_params(width=0.219*2.83465, which='minor')
ax.yaxis.set_tick_params(width=0.219*2.83465, which='minor')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
### find remaining length of iron that we can model as a slab.

altered_Fe_len = Fe_thickness - (medpp[0][pos_to_start]-Fe_FeOx_mdpoint)
```

## Now we do the refnx comparison

```python
from refnx.dataset import ReflectDataset
from refnx.analysis import Transform, CurveFitter, Objective, GlobalObjective, Parameter
from refnx.reflect import SLD, ReflectModel, MixedReflectModel, MixedSlab, Slab, Structure
from MixedLayerSlab import MixedMagSlabs2
import refnx
import scipy
```

```python
print(f'''
refnx: {refnx.version.version}
scipy: {scipy.version.version}
numpy: {np.version.version}''')
```

```python
#load data from .txt files - this can be done smarter but works as a test.

file_path_GMOddod25 = '61167_69.txt'
file_path_GMOCMdod25 = '61203_05.txt'
file_path_GMOhdod25 = '61215_17.txt'

#Define datasets with data loaded

P1_GMO_ddod_25 = ReflectDataset(file_path_GMOddod25)
P1_GMO_CMdod_25 = ReflectDataset(file_path_GMOCMdod25)
P1_GMO_hdod_25 = ReflectDataset(file_path_GMOhdod25)
```

```python
#now define some SLDs we will use...

#LAYERS#
Si = SLD(2.07, name='Si')
SiO2 = SLD(3.47, name='SiO2')
Fe = SLD(7.914021, name='Fe')

#SOLVENTS#
hdod_SLD = SLD(np.average(hd_comp_sldn[0:15]), name='hdod')
ddod_SLD = SLD(np.average(dd_comp_sldn[0:15]), name='ddod')
CMdod_SLD = SLD(np.average(CMdod_comp_sldn[0:15]), name='CMdod')
```

```python
"""
Here we define the material layers at the interface.
"""

#Create slabs for model

##SiO2##

SiO2_lay = SiO2(SiO2_thickness, 3)

SiO2_lay.thick.setp(vary=False, bounds=(1, 25))

##Fe##
# share the same Fe thickness, SiO2-Fe roughness - we specify variation of parameters here.
SiO2_Fe_R = Parameter(SiO2_Fe_roughness, 'SiO2_Fe_R', vary=False, bounds=(1, 10))
Fe.real.setp(vary=False, bounds=(7.5, 8.1))
Fe_thick = Parameter(altered_Fe_len, 'Fe_t', vary=False, bounds=(170, 210))
Fe_m = Parameter((((7.914021*1E-6)/0.0000945)*2.091443E+00*0.00002699*1E6), 'Fe_m', vary=False)

FeUp_lay = MixedMagSlabs2.USM_nomagmom_Slab(Fe_thick, Fe, Fe_m, SiO2_Fe_R)
FeDown_lay = MixedMagSlabs2.DSM_nomagmom_Slab(Fe_thick, Fe, Fe_m, SiO2_Fe_R)

hdod_GMO25_lay = hdod_SLD(0, 0)
ddod_GMO25_lay = ddod_SLD(0, 0)
CMdod_GMO25_lay = CMdod_SLD(0, 0)
```

```python
ddlayers_pp = []
for k, j in enumerate(np.flip(dd_comp_sldn)):
    Lay = Slab(thick=0.5, sld=j+np.flip(comp_sldm)[k], rough=0.00)
    ddlayers_pp.append(Lay)
    
ddlayers_mm = []
for k, j in enumerate(np.flip(dd_comp_sldn)):
    Lay = Slab(thick=0.5, sld=j-np.flip(comp_sldm)[k], rough=0.00)
    ddlayers_mm.append(Lay)

CMdodlayers_pp = []
for k, j in enumerate(np.flip(CMdod_comp_sldn)):
    Lay = Slab(thick=0.5, sld=j+np.flip(comp_sldm)[k], rough=0.00)
    CMdodlayers_pp.append(Lay)
    
CMdodlayers_mm = []
for k, j in enumerate(np.flip(CMdod_comp_sldn)):
    Lay = Slab(thick=0.5, sld=j-np.flip(comp_sldm)[k], rough=0.00)
    CMdodlayers_mm.append(Lay)

hdlayers_pp = []
for k, j in enumerate(np.flip(hd_comp_sldn)):
    Lay = Slab(thick=0.5, sld=j+np.flip(comp_sldm)[k], rough=0.00)
    hdlayers_pp.append(Lay)
    
hdlayers_mm = []
for k, j in enumerate(np.flip(hd_comp_sldn)):
    Lay = Slab(thick=0.5, sld=j-np.flip(comp_sldm)[k], rough=0.00)
    hdlayers_mm.append(Lay)
```

```python
stack_dd_pp = Structure(components=ddlayers_pp)
stack_dd_mm = Structure(components=ddlayers_mm)
stack_hd_pp = Structure(components=hdlayers_pp)
stack_hd_mm = Structure(components=hdlayers_mm)
stack_CMdod_pp = Structure(components=CMdodlayers_pp)
stack_CMdod_mm = Structure(components=CMdodlayers_mm)
```

```python
P1_GMO25ddod_U = Si | SiO2_lay | FeUp_lay | stack_dd_pp #| ddod_GMO25_lay
P1_GMO25ddod_D = Si | SiO2_lay | FeDown_lay | stack_dd_mm #| ddod_GMO25_lay

P1_GMO25CMdod_U = Si | SiO2_lay | FeUp_lay | stack_CMdod_pp #| CMdod_GMO25_lay
P1_GMO25CMdod_D = Si | SiO2_lay | FeDown_lay | stack_CMdod_mm #| CMdod_GMO25_lay

P1_GMO25hdod_U = Si | SiO2_lay | FeUp_lay | stack_hd_pp #| hdod_GMO25_lay
P1_GMO25hdod_D = Si | SiO2_lay | FeDown_lay | stack_hd_mm #| hdod_GMO25_lay
```

```python
plt.plot(*P1_GMO25CMdod_U.sld_profile())
plt.plot(*P1_GMO25CMdod_D.sld_profile())
```

```python
#now define the models

#these are the scale and background parameters
#intensities
P1_GMO25dd_int = Parameter((9.483492E-01/2), 'P1_GMO25dd_int', vary=False, bounds=(0.5*0.8, 0.5*1.2))
P1_GMO25CMd_int = Parameter((9.398290E-01/2), 'P1_GMO25CMd_int', vary=False, bounds=(0.5*0.8, 0.5*1.2))
P1_GMO25hd_int = Parameter((9.470335E-01/2), 'P1_GMO25hd_int', vary=False, bounds=(0.5*0.8, 0.5*1.2))
#background
P1_GMO25dd_bkg = Parameter(4.850123E-06, 'P1_GMO25dd_bkg', vary=False, bounds=(1e-7, 2e-5))
P1_GMO25CMd_bkg = Parameter(9.209948E-06, 'P1_GMO25CMd_bkg', vary=False, bounds=(1e-7, 2e-5))
P1_GMO25hd_bkg = Parameter(6.061751E-06, 'P1_GMO25hd_bkg', vary=False, bounds=(1e-7, 2e-5))

#define adv_layer interface objective & global objective. These adv layers have their own SLD.
#models
P1GMOddod25_M = MixedReflectModel((P1_GMO25ddod_U, P1_GMO25ddod_D), scales=(P1_GMO25dd_int, P1_GMO25dd_int), bkg=P1_GMO25dd_bkg, dq=4.7096)
P1GMOCMdod25_M = MixedReflectModel((P1_GMO25CMdod_U, P1_GMO25CMdod_D), scales=(P1_GMO25CMd_int, P1_GMO25CMd_int), bkg=P1_GMO25CMd_bkg, dq=4.7096)
P1GMOhdod25_M = MixedReflectModel((P1_GMO25hdod_U, P1_GMO25hdod_D), scales=(P1_GMO25hd_int, P1_GMO25hd_int), bkg=P1_GMO25hd_bkg, dq=4.7096)
#objectives
P1GMOddod25_obj = Objective(P1GMOddod25_M, P1_GMO_ddod_25)
P1GMOCMdod25_obj = Objective(P1GMOCMdod25_M, P1_GMO_CMdod_25)
P1GMOhdod25_obj = Objective(P1GMOhdod25_M, P1_GMO_hdod_25)
#global obj
glob_objective = GlobalObjective([P1GMOddod25_obj, P1GMOCMdod25_obj, P1GMOhdod25_obj])
```

```python
#calculate reduced / normalised chi2. Note varying_parameters length should be 0.
chi2 = glob_objective.chisqr()
chi2points = chi2/(glob_objective.npoints)
```

```python
### median structure of fitted models ### 

#LAYERS#
Si_med = SLD(2.07, name='Si_med')
SiO2_med = SLD(3.47, name='SiO2_med')
Fe_med = SLD(7.914021E+00, name='Fe_med')
FeOx_med = SLD(5.637175E+00, name='FeOx_med')
GMO_med = SLD(0.21, name='GMO_med')

#SOLVENTS#
hdod_SLD_med = SLD(-0.462, name='hdod_med')
ddod_med = SLD(solvent_SLD, name='ddod_med')
CMdod_med = SLD(1.926268E+00, name='CMdod_med')

##SiO2##

SiO2_lay_med = SiO2_med(SiO2_thickness, 3)

SiO2_lay_med.thick.setp(vary=False, bounds=(1, 25))

##Fe##
# share the same Fe thickness, SiO2-Fe roughness - we specify variation of parameters here.
SiO2_Fe_R_med = Parameter(SiO2_Fe_roughness, 'SiO2_Fe_R_med', vary=False, bounds=(1, 15))
Fe_med.real.setp(vary=False, bounds=(7.5, 8.1))
Fe_thick_med = Parameter(Fe_thickness, 'Fe_t_med', vary=False, bounds=(150, 250))
Fe_magmom_med = Parameter(2.091443E+00, 'Fe_magmom_med', vary=False, bounds=(1.9, 2.2))
Fe_ScattLen_med = Parameter(0.0000945, 'Fe_SL_med', vary=False)

FeUp_lay_med = MixedMagSlabs2.USM_Slab(Fe_thick_med, Fe_med, Fe_magmom_med, Fe_ScattLen_med, SiO2_Fe_R_med)
FeDown_lay_med = MixedMagSlabs2.DSM_Slab(Fe_thick_med, Fe_med, Fe_magmom_med, Fe_ScattLen_med, SiO2_Fe_R_med)

##FeOx##
Fe_FeOx_R_med = Parameter(Fe_FeOx_roughness, 'FeOx_Fe_R_med', vary=False, bounds=(1, 15))
FeOx_med.real.setp(vary=True, bounds=(5.0, 7.2))
FeOx_thick_med = Parameter(FeOx_thickness, 'FeOx_t_med', vary=False, bounds=(20, 60))
FeOx_m_med = Parameter(1.148236E-01, 'FeOx_m_med', vary=False, bounds=(0, 1.3))

FeOxUp_lay_med = MixedMagSlabs2.USM_nomagmom_Slab(FeOx_thick_med, FeOx_med, FeOx_m_med, Fe_FeOx_R_med)
FeOxDown_lay_med = MixedMagSlabs2.DSM_nomagmom_Slab(FeOx_thick_med, FeOx_med, FeOx_m_med, Fe_FeOx_R_med)

### GMO

GMO_lay_med = GMO_med(GMO_thickness, 4.796275E+00)
GMO_lay_med.vfsolv.setp(GMO_solvation)

GMO_R = Parameter(GMO_solvent_roughness, 'GMO_R_med', vary=False, bounds=(1, 15))

### SOLVENT PARAMS FOR ADV_LAYER SOLVENT
#setup the solvent layers

hdod_lay_med = hdod_SLD_med(0, GMO_R)
ddod_lay_med = ddod_med(0, GMO_R)
CMdod_lay_med = CMdod_med(0, GMO_R)

GMO_dd_U = Si_med | SiO2_lay_med | FeUp_lay_med | FeOxUp_lay_med | GMO_lay_med | ddod_lay_med
GMO_dd_D = Si_med | SiO2_lay_med | FeDown_lay_med | FeOxDown_lay_med | GMO_lay_med | ddod_lay_med
GMO_CMd_U = Si_med | SiO2_lay_med | FeUp_lay_med | FeOxUp_lay_med | GMO_lay_med | CMdod_lay_med
GMO_CMd_D = Si_med | SiO2_lay_med | FeDown_lay_med | FeOxDown_lay_med | GMO_lay_med | CMdod_lay_med
GMO_hd_U = Si_med | SiO2_lay_med | FeUp_lay_med | FeOxUp_lay_med | GMO_lay_med | hdod_lay_med
GMO_hd_D = Si_med | SiO2_lay_med | FeDown_lay_med | FeOxDown_lay_med | GMO_lay_med | hdod_lay_med
```
