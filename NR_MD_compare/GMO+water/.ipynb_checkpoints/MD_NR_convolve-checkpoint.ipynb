```python
#load fitted objective.

import sys
import os

sys.path.append(os.getcwd())

from refnx.dataset import ReflectDataset
from refnx.analysis import Transform, CurveFitter, Objective, GlobalObjective, Parameter, autocorrelation_chain, process_chain, integrated_time
from refnx.reflect import SLD, ReflectModel, MixedReflectModel, MixedSlab, Slab, Structure
import MixedMagSlabs2
from vfp_M2 import VFP
import refnx
import scipy
import matplotlib.pyplot as plt
import numpy as np
import pickle

class LogpExtra(object):
    def __init__(self, pars):
        # we'll store the parameters and objective in this object
        # this will be necessary for pickling in the future
        self.pars = pars

    def __call__(self, model, data):
        if (float(self.pars[0]-2*self.pars[1]) >= 0 and float(self.pars[2]-2*self.pars[3]) >= 0 and float(self.pars[6]) >= 0 and float(self.pars[6]) <= 1 
        and float(self.pars[4]-2*self.pars[5]) >= 0 and float(self.pars[7]) >= 0 and float(self.pars[7]) <= 1):
            return 0
        return -1E15
        
with open(os.path.join(os.getcwd(), 'M2_FIT_GO.pkl'),"rb") as f:
    GO = pickle.load(f)
```

```python
#we want the SLD profile of the underlying substrate (which does not depend on the solvent contrast nor GMO+water contrast)
#create nuclear and magnetic sld profiles from the Si, SiO2, Fe and FeOx layers...

Fe_sld_n = GO.objectives[0].varying_parameters()[7]
Fe_magmom = GO.objectives[0].varying_parameters()[8]
FeOx_sld_n = GO.objectives[0].varying_parameters()[10]
FeOx_sld_m = GO.objectives[0].varying_parameters()[12]
#first lets create a list of the Si, SiO2, FeUp, FeDown, FeOxUp & FeOxDown slds.
SLDs = [2.07, 
        3.47, 
        Fe_sld_n+(((Fe_sld_n*1E-6)/(9.45*1E-5))*Fe_magmom*2.699*1E-5*1E6),
        Fe_sld_n-(((Fe_sld_n*1E-6)/(9.45*1E-5))*Fe_magmom*2.699*1E-5*1E6),
        FeOx_sld_n+FeOx_sld_m,
        FeOx_sld_n-FeOx_sld_m]

up_SLDs = np.array([float(SLDs[i]) for i in [0, 1, 2, 4]])
up_each_SLD = median_volume_fracs[0:4].T*up_SLDs
up_tot_SLD = np.sum(up_each_SLD, 1)

down_SLDs = np.array([float(SLDs[i]) for i in [0, 1, 3, 5]])
down_each_SLD = median_volume_fracs[0:4].T*down_SLDs
down_tot_SLD = np.sum(down_each_SLD, 1)

nuclear_SLD = (up_tot_SLD+down_tot_SLD)/2
magnetic_SLD = (up_tot_SLD-down_tot_SLD)/2
```

```python
#load MD SLD profiles.

df_vf=pd.read_csv(os.path.join(os.getcwd(), 'gmo_1_water_1_sld.csv'), sep=',' ,header=[0])
```

```python
#create arrays from csv data.

z = df_vf.values.T[0]
GMO = df_vf.values.T[1]*1E6
dodecane = df_vf.values.T[2]*1E6
H2O = df_vf.values.T[3]*1E6
D2O = df_vf.values.T[4]*1E6
```

```python
#modify the dodecane SLDs to match experimental SLD.

dd_d2o_mod = ((float(GO.objectives[0].varying_parameters()[-5]))/(np.average(dodecane[-16:-1])))*dodecane #scale dd sld
dd_h2o_mod = ((float(GO.objectives[0].varying_parameters()[-4]))/(np.average(dodecane[-16:-1])))*dodecane #scale dd sld
hd_mod = ((-0.462)/(np.average(dodecane[-16:-1])))*dodecane #scale hd sld
```

```python
#extend the SLD profile of each component by taking last 15 data points in each array and repeating 10 times.
#this step is required so that the convolved data reaches a steady bulk SLD for solvent.

dd_d2o_mod = np.append(dd_d2o_mod, np.tile(dd_d2o_mod[-15:-1], 10))
dd_h2o_mod = np.append(dd_h2o_mod, np.tile(dd_h2o_mod[-15:-1], 10))
hd_mod = np.append(hd_mod, np.tile(hd_mod[-15:-1], 10))
GMO = np.append(GMO, np.tile(GMO[-15:-1], 10))
H2O = np.append(H2O, np.tile(H2O[-15:-1], 10))
D2O = np.append(D2O, np.tile(D2O[-15:-1], 10))

#extend the distance axis too.
z2 = np.append(z, z[-1]+np.cumsum(0.5*np.ones(dd_d2o_mod.shape[0]-z.shape[0])))

#then we can change the spacing of z2 for interpolation (improve convolution approximate to original SLD curve).
#interpolate to 0.1 Å.
z2_interp = np.linspace(z2[0], np.round(z2[-1]*2)/2, int((np.round(z2[-1]*2)/2 - z2[0])/0.1)+1)
```

```python
D2O_GMO_dod = D2O+GMO+dd_d2o_mod
H2O_GMO_dod = H2O+GMO+dd_h2o_mod
D2O_GMO_hdod = D2O+GMO+hd_mod

plt.plot(z2, D2O_GMO_dod)
plt.plot(z2, H2O_GMO_dod)
plt.plot(z2, D2O_GMO_hdod)

plt.xlabel(r'Distance from FeO$_{x}$ / $\mathrm{\AA{}}$')
plt.ylabel(r'$\beta$ / $\mathrm{\AA{}}^{-2}\times10^{-6}$')
```

```python
#find distances of volume fractions.
zed = GO.objectives[0].model.structures[0].components[1].get_x_and_y_scatter()[0]

#match spacing of zed to that of z2_interp.
zed2 = np.linspace(zed[0], np.round(zed[-1]*2)/2, int((np.round(zed[-1]*2)/2 - zed[0])/0.1)+1)
```

```python
#interpolate Rui's SLD profiles.

D2O_GMO_dod_interp = np.interp(z2_interp, z2, D2O_GMO_dod)
H2O_GMO_dod_interp = np.interp(z2_interp, z2, H2O_GMO_dod)
D2O_GMO_hdod_interp = np.interp(z2_interp, z2, D2O_GMO_hdod)
```

```python
#check original (extended) SLD profile and interpolated appear similar.

%matplotlib inline
plt.plot(z2, D2O_GMO_dod, zorder=5)
plt.plot(z2_interp, D2O_GMO_dod_interp, marker='.')
plt.xlabel(r'Distance from FeO$_{x}$ / $\mathrm{\AA{}}$')
plt.ylabel(r'$\beta$ / $\mathrm{\AA{}}^{-2}\times10^{-6}$')
```

```python
#interpolate SLDn, SLDm and the vf profile of Si, SiO2, Fe, FeOx...

nuclear_SLD_interp = np.interp(zed2, zed, nuclear_SLD)
magnetic_SLD_interp = np.interp(zed2, zed, magnetic_SLD)

vf_interp = np.interp(zed2, zed, median_volume_fracs[0]+median_volume_fracs[1]+median_volume_fracs[2]+median_volume_fracs[3])
conv_by = np.gradient(1-vf_interp) #convolve the SLD profile by 1-vf_interp
```

```python
# we want to convolute Rui's SLD profiles by 1-vf of Si, SiO2, Fe, FeOx...

plt.plot(GO.objectives[0].model.structures[0].components[1].get_x_and_y_scatter()[0], median_volume_fracs[0]+median_volume_fracs[1]+median_volume_fracs[2]+median_volume_fracs[3])
plt.plot(zed2, 1-vf_interp)

plt.xlabel(r'Distance from Interface / $\mathrm{\AA{}}$')
plt.ylabel(r'Volume Fraction')
```

```python
#check SLDs look similar before and after interpolation

plt.plot(zed, magnetic_SLD, label='Magnetic')
plt.plot(zed, nuclear_SLD, label='Nuclear')

plt.plot(zed2, magnetic_SLD_interp, linestyle='--', color='k', label='After Interpolation')
plt.plot(zed2, nuclear_SLD_interp, linestyle='--', color='k')

plt.legend(frameon=False)

plt.xlabel(r'Distance from Interface / $\mathrm{\AA{}}$')
plt.ylabel(r'$\beta$ / $\mathrm{\AA{}}^{-2}\times10^{-6}$')
```

```python
#do the convolution here...

GMO_D2O_dd_conv = np.convolve(conv_by, D2O_GMO_dod_interp, mode='full')
GMO_H2O_dd_conv = np.convolve(conv_by, H2O_GMO_dod_interp, mode='full')
GMO_D2O_hd_conv = np.convolve(conv_by, D2O_GMO_hdod_interp, mode='full')
```

```python
#MD SLD profiles after convolution (note not plotted with distances)
#note that the curve at >~ 4000 is a boundary effect - we need to cut off the SLD profile before this is reached.

plt.plot(GMO_D2O_dd_conv, label='D2O_dd')
plt.plot(GMO_H2O_dd_conv, label='H2O_dd')
plt.plot(GMO_D2O_hd_conv, label='D2O_hd')
plt.legend()
```

```python
#extend the Si, SiO2, Fe, FeOx SLD lists to be as long as the convolved datasets.

SLDn_mod = np.append(nuclear_SLD_interp, np.zeros(GMO_D2O_dd_conv.shape[0]-nuclear_SLD_interp.shape[0]))
SLDm_mod = np.append(magnetic_SLD_interp, np.zeros(GMO_D2O_dd_conv.shape[0]-nuclear_SLD_interp.shape[0]))
```

```python
#now we need a distance array as long as the convolved datasets.

zed3 = np.append(zed2, zed2[-1]+np.cumsum(0.1*np.ones(SLDn_mod.shape[0]-zed2.shape[0])))
```

```python
#Rui's SLD profiles after convolution as function of distance.

plt.plot(zed3, GMO_D2O_dd_conv, label='D2O_dd')
plt.plot(zed3, GMO_H2O_dd_conv, label='H2O_dd')
plt.plot(zed3, GMO_D2O_hd_conv, label='D2O_hd')
plt.legend()

plt.xlabel(r'Distance from Interface / $\mathrm{\AA{}}$')
plt.ylabel(r'$\beta$ / $\mathrm{\AA{}}^{-2}\times10^{-6}$')
```

```python
#See the addition of convolved SLD profiles with the interpolated SLD n profiles.

plt.plot(zed3, GMO_D2O_dd_conv+SLDn_mod, label='D2O_dd')
plt.plot(zed3, GMO_H2O_dd_conv+SLDn_mod, label='H2O_dd')
plt.plot(zed3, GMO_D2O_hd_conv+SLDn_mod, label='D2O_hd')

plt.plot(zed3, SLDm_mod, label='Magnetic')

plt.xlabel(r'Distance from Interface / $\mathrm{\AA{}}$')
plt.ylabel(r'$\beta$ / $\mathrm{\AA{}}^{-2}\times10^{-6}$')
plt.legend(frameon=False, loc='lower left', bbox_to_anchor=(0.2, 0.2))
```

```python
#create up and down data:

dd_d2o_up_conv = GMO_D2O_dd_conv+SLDn_mod+SLDm_mod
dd_d2o_down_conv = GMO_D2O_dd_conv+SLDn_mod-SLDm_mod

dd_h2o_up_conv = GMO_H2O_dd_conv+SLDn_mod+SLDm_mod
dd_h2o_down_conv = GMO_H2O_dd_conv+SLDn_mod-SLDm_mod

hd_d2o_up_conv = GMO_D2O_hd_conv+SLDn_mod+SLDm_mod
hd_d2o_down_conv = GMO_D2O_hd_conv+SLDn_mod-SLDm_mod

plt.plot(zed3, dd_d2o_up_conv)
plt.plot(zed3[:700*5], dd_d2o_up_conv[:700*5])
plt.plot(zed3[:700*5], hd_d2o_up_conv[:700*5])
```

```python
#load data from .txt files - this can be done smarter but works as a test.

file_path_GMO_H2O_ddod25 = '61115_17.txt'
file_path_GMO_D2O_ddod25 = '61146_48.txt'
file_path_GMO_D2O_hdod25 = '61150_52.txt'

#Define datasets with data loaded

L1_GMO_H2O_ddod_25 = ReflectDataset(file_path_GMO_H2O_ddod25)
L1_GMO_D2O_ddod_25 = ReflectDataset(file_path_GMO_D2O_ddod25)
L1_GMO_D2O_hdod_25 = ReflectDataset(file_path_GMO_D2O_hdod25)
```

```python
z_diff = 0.1
# or we can just build from vf profiles?
ddlayers_pp = []
for i in dd_d2o_up_conv[:700*5]:
    Lay = Slab(thick=z_diff, sld=i, rough=0)
    ddlayers_pp.append(Lay)
    
ddlayers_mm = []
for i in dd_d2o_down_conv[:700*5]:
    Lay = Slab(thick=z_diff, sld=i, rough=0)
    ddlayers_mm.append(Lay)
    
ddhlayers_pp = []
for i in dd_h2o_up_conv[:700*5]:
    Lay = Slab(thick=z_diff, sld=i, rough=0)
    ddhlayers_pp.append(Lay)
    
ddhlayers_mm = []
for i in dd_h2o_down_conv[:700*5]:
    Lay = Slab(thick=z_diff, sld=i, rough=0)
    ddhlayers_mm.append(Lay)
    
hdlayers_pp = []
for i in hd_d2o_up_conv[:700*5]:
    Lay = Slab(thick=z_diff, sld=i, rough=0)
    hdlayers_pp.append(Lay)
    
hdlayers_mm = []
for i in hd_d2o_down_conv[:700*5]:
    Lay = Slab(thick=z_diff, sld=i, rough=0)
    hdlayers_mm.append(Lay)
```

```python
stack_dd_pp_alt = Structure(components=ddlayers_pp)
stack_dd_mm_alt = Structure(components=ddlayers_mm)

stack_ddh_pp_alt = Structure(components=ddhlayers_pp)
stack_ddh_mm_alt = Structure(components=ddhlayers_mm)

stack_hd_pp_alt = Structure(components=hdlayers_pp)
stack_hd_mm_alt = Structure(components=hdlayers_mm)
```

```python
L1_GMO25dd_d2o_U_alt = stack_dd_pp_alt
L1_GMO25dd_d2o_D_alt = stack_dd_mm_alt

L1_GMO25dd_h2o_U_alt = stack_ddh_pp_alt
L1_GMO25dd_h2o_D_alt = stack_ddh_mm_alt

L1_GMO25hd_d2o_U_alt = stack_hd_pp_alt
L1_GMO25hd_d2o_D_alt = stack_hd_mm_alt
```

```python
#now define the models

#these are the scale and background parameters
#intensities
L1_GMO25dd_d2o_int = Parameter(GO.varying_parameters()[-6], 'L1_GMO25dd_d2o_int', vary=False, bounds=(0.5*0.8, 0.5*1.2))
L1_GMO25dd_h2o_int = Parameter(GO.varying_parameters()[-4], 'L1_GMO25dd_h2o_int', vary=False, bounds=(0.5*0.8, 0.5*1.2))
L1_GMO25hd_d2o_int = Parameter(GO.varying_parameters()[-2], 'L1_GMO25hd_d2o_int', vary=False, bounds=(0.5*0.8, 0.5*1.2))
#background
L1_GMO25dd_d2o_bkg = Parameter(GO.varying_parameters()[-5], 'L1_GMO25dd_d2o_bkg', vary=False, bounds=(1e-7, 2e-5))
L1_GMO25dd_h2o_bkg = Parameter(GO.varying_parameters()[-3], 'L1_GMO25dd_h2o_bkg', vary=False, bounds=(1e-7, 2e-5))
L1_GMO25hd_d2o_bkg = Parameter(GO.varying_parameters()[-1], 'L1_GMO25hd_d2o_bkg', vary=False, bounds=(1e-7, 2e-5))

#define adv_layer interface objective & global objective. These adv layers have their own SLD.
#models
L1_GMO25dd_d2o_M_conv = MixedReflectModel((L1_GMO25dd_d2o_U_alt, L1_GMO25dd_d2o_D_alt), scales=(L1_GMO25dd_d2o_int, L1_GMO25dd_d2o_int), bkg=L1_GMO25dd_d2o_bkg, dq=7.064460135)
L1_GMO25dd_h2o_M_conv = MixedReflectModel((L1_GMO25dd_h2o_U_alt, L1_GMO25dd_h2o_D_alt), scales=(L1_GMO25dd_h2o_int, L1_GMO25dd_h2o_int), bkg=L1_GMO25dd_h2o_bkg, dq=7.064460135)
L1_GMO25hd_d2o_M_conv = MixedReflectModel((L1_GMO25hd_d2o_U_alt, L1_GMO25hd_d2o_D_alt), scales=(L1_GMO25hd_d2o_int, L1_GMO25hd_d2o_int), bkg=L1_GMO25hd_d2o_bkg, dq=7.064460135)
#objectives
L1GMOdd_d2o_25_obj_conv = Objective(L1_GMO25dd_d2o_M_conv, L1_GMO_D2O_ddod_25)
L1GMOdd_h2o_25_obj_conv = Objective(L1_GMO25dd_h2o_M_conv, L1_GMO_H2O_ddod_25)
L1GMOhd_d2o_25_obj_conv = Objective(L1_GMO25hd_d2o_M_conv, L1_GMO_D2O_hdod_25)
#global obj
glob_objective_conv = GlobalObjective([L1GMOdd_d2o_25_obj_conv, L1GMOdd_h2o_25_obj_conv, L1GMOhd_d2o_25_obj_conv])
```
