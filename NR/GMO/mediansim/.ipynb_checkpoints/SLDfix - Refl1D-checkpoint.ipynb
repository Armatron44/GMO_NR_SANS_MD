```python
#Author A. Armstrong. V1 - 9-4-2019
#Converts SLD profiles of S/L systems generated by Refl1D. 
#This sets the correct posistion for zero - halfway between SiO2 and Si

import numpy as np
from tkinter import *
import tkinter.filedialog as tk
from pathlib import Path
import os, sys

root = Tk() # a tkinter window is created - this is is known as root.

SLD2ndlayer = input('SLD of layer after sub x 1E-6? e.g SiO2 value is 3.467 ') #asks user for SLD value of layer after sub
no2 = float(SLD2ndlayer) #converts the string input into a floating point value
SLDsublayer = input('SLD of sub layer x 1E-6? e.g Si value is 2.072 ') 
no1 = float(SLDsublayer) 
halfway = ((no2 - no1)/2) + no1 #calculates the SLD value of the true interface that is halfway between the two layers.

file_list = [] #initialise a list called file_list
files = tk.askopenfilenames(title='Choose files') #variable called files opens a tkinter dialog to choose files.
file_list = list(files) #fills the file_list list with files selected in the tkinter dialog

root.destroy() # destroys all widgets and exits the mainloop. Any code after the call will run but attempts to access widgets later in the code will not work.

l = 0 #vairable called l initialised at 0. This is used to count the number of iterations through the main for loop.

for files in file_list: #for the number of files selected from the tk dialog. This is the main for loop.
    with open(files) as f: #defines the list 'files' as f and opens each file in the list and then closes it automatically
        
        #This block initialises a load of lists and variables that are used in the main for loop.
        a = [] 
        b = []
        c = []
        k = 0
        u = 0
        newdistance = []
        results = []
        Nearinterface = []
        fullSLD= []
        tkinterpath = []
        
        for line in f: #for the number of lines in an individual file...
            if not '#' in line: #ignoring the lines that contain #
                a0,b0,c0 = line.strip().split() #strip and split each line by removing spaces before & after string and removing space between values in line - () means space
        
                a.append(float(a0)) #take the value of a0 stripped from above and put it into the a list. 
                b.append(float(b0))
                c.append(float(c0))
    a2 = []
    b2 = []
    for le, ke in enumerate(a):
        if le % 5 == 0:
            a2.append(a[le])
            b2.append(b[le])
    for number in b2: #For every number in c...
        results.append(number - halfway) #take the variable halfway from each value in c and put this value in results.
    Nearinterface = results[-200:-1] #Refine the results list down to the interface
    UpperSLD = min([i for i in Nearinterface if i>0]) #Find the minimum posistive value from the nearinterface list 
    us = results.index(UpperSLD) # us is a variable that is the index value of where the value of upperSLD is in the list
    DeltaPoints = ((a2[us] - a2[us+1])**2)**(1/2) #the magnitude of the difference between the distance values
   
    DeltaSLD = b2[us] - b2[us+1] #the value of difference between the SLD values
    DiffSLDratio = (b2[us] - halfway) / DeltaSLD #finds ratio of where interface lies between the two data points.
    USnew = DeltaPoints * DiffSLDratio #uses ratio to set the correct value of distance from interface for 1 point.
    Uppernd = USnew + DeltaPoints * us #first value in distance list is set to correct value and for loop used below used to find others.
    
    for g, h in enumerate(b2): #for numbers in list x
        newdistance.append(Uppernd - g*DeltaPoints) #subtracts an increasing factor of 0.5 from top value in list.
    NuclearSLD = b2
    tkinterpath = file_list[l] #paths for each file are seperated by using the index l
    directorypath = "/".join(tkinterpath.split("/")[:-1]) #variable for directory path made by splitting string and taking last sub-string in list.
    filename = (tkinterpath.split("/")[-1]) #variable for filename made by splitting string and taking 13th substring
    newfilename = ".".join(filename.split(".")[:-1]) + "-corrected.txt" #filename then has .dat removed and is concatenated with -corrected.txt
    l += 1
    os.chdir(directorypath) #using the os package, we select the directory to save new file in as the directorypath variable.
    with open(newfilename, "w") as w: #creates a new file with the name specified by variable newfilename.
        for u, number in enumerate(newdistance): #for every number in the new distance list
            w.write(str(newdistance[u]) + " " + str(NuclearSLD[u]) + "\n") #values in both lists are written with space between them as strings.
            
    w.close() #this ensures that the file opened and written into is then closed.
```
